# Baseline

```python
df_base=pd.read_csv("times/base_benchmark.csv", sep=",")
#change the type of first column to int
df_base["Query"]=df_base["Query"].astype(int)
```

```python
#mean of the query 1 values, keep only the first 3 digits
mean_query1_base=round(df_base[df_base["Query"]==1]["Execution Time [ms]"].mean(),3)
mean_query10_base=round(df_base[df_base["Query"]==10]["Execution Time [ms]"].mean(),3)
mean_query14_base=round(df_base[df_base["Query"]==14]["Execution Time [ms]"].mean(),3)
mean_query17_base=round(df_base[df_base["Query"]==17]["Execution Time [ms]"].mean(),3)

#std of the query 1 values
std_query1_base=round(df_base[df_base["Query"]==1]["Execution Time [ms]"].std(),3)
std_query10_base=round(df_base[df_base["Query"]==10]["Execution Time [ms]"].std(),3)
std_query14_base=round(df_base[df_base["Query"]==14]["Execution Time [ms]"].std(),3)
std_query17_base=round(df_base[df_base["Query"]==17]["Execution Time [ms]"].std(),3)

#print the statistics
print("Query, mean, std")
print(f"Query 1     ,{mean_query1_base}                ,{std_query1_base}")
print(f"Query 10    ,{mean_query10_base}               ,{std_query10_base}")
print(f"Query 14    ,{mean_query14_base}               ,{std_query14_base}")
print(f"Query 17    ,{mean_query17_base}               ,{std_query17_base}")
```

```python
plt.figure(figsize=(10, 6))

sns.violinplot(x='Query', y='Execution Time [ms]', data=df_base, palette=colors, hue='Query')

plt.title('Baseline')
plt.xlabel('Query Number')
plt.ylabel('Execution Time [ms]')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_double_barplot(df : pd.DataFrame, save = False, name=None):
    # Plot execution time with error bars

    x = np.arange(len(df['Method']))  # the label locations

    fig, ax1 = plt.subplots(figsize=(10, 6))

    ax1.bar(x - 0.2, df["Execution Time [ms]"], width=0.4, yerr=df['Std'], capsize=5, color='b', alpha=0.6, label='Execution Time')
    ax1.set_xlabel('Database')
    ax1.set_ylabel('Execution Time (s)', color='b')
    ax1.set_xticks(x)
    ax1.set_xticklabels(df['Method'])
    ax1.tick_params(axis='y', labelcolor='b')

    # Plot database dimensions on the same x-axis
    ax2 = ax1.twinx()
    ax2.bar(x + 0.2, df["db_dimension [GB]"], width=0.4, color='darkorange', alpha=0.6, label='Database Dimension')
    ax2.set_ylabel('Database Dimension (MB)', color='darkorange')
    ax2.tick_params(axis='y', labelcolor='darkorange')

    # Add a legend
    fig.tight_layout()
    fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

    #plt.title('Execution Time and Database Dimension')

    if save:
        plt.savefig(f'plots/double_barplot_{name}.png', dpi=300, bbox_inches='tight')
```
