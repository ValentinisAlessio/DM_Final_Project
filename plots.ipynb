```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colors="Set3"
```

# Baseline

```python
df_base=pd.read_csv("times/base_benchmark.csv", sep=",")
#change the type of first column to int
df_base["Query"]=df_base["Query"].astype(int)
```

```python
plt.figure(figsize=(10, 6))

sns.violinplot(x='Query', y='Execution Time [ms]', data=df_base, palette=colors, hue='Query')

plt.title('Baseline')
plt.xlabel('Query Number')
plt.ylabel('Execution Time [ms]')
plt.show()
```

# Index

```python
df_index=pd.read_csv("times/index_benchmark.csv", sep=",")
#change the type of first column to int
df_index["Query"]=df_index["Query"].astype(int)
```

```python
plt.figure(figsize=(10, 6))

sns.violinplot(x='Query', y='Execution Time [ms]', data=df_index, palette=colors, hue='Query')

plt.title('Index')
plt.xlabel('Query Number')
plt.ylabel('Execution Time [ms]')
#plt.ylim(0, 700000)

plt.show()
```

# Materialized

# Fragmentation

```python
df_frag=pd.read_csv("times/fragmentation_benchmark.csv", sep=",")
#change the type of first column to int
df_frag["Query"]=df_frag["Query"].astype(int)
```

```python
plt.figure(figsize=(10, 6))

sns.violinplot(x='Query', y='Execution Time [ms]', data=df_frag, palette=colors, hue='Query')

plt.title('Fragmentation')
plt.xlabel('Query Number')
plt.ylabel('Execution Time [ms]')
plt.show()
```

# Plot per query

## Query 1

```python
#create a dataframe with contains all element of query 1
df_base_q1=df_base[df_base['Query']==1]
df_index_q1=df_index[df_index['Query']==1]
df_frag_q1=df_frag[df_frag['Query']==1]
#merge in a single datagrame which has 2 columns: Execution Time [ms] and Method
df_q1=pd.concat([df_base_q1, df_index_q1, df_frag_q1], ignore_index=True)
df_q1['Method']=['Baseline']*len(df_base_q1)+['Index']*len(df_index_q1)+['Fragmentation']*len(df_frag_q1)
#frop first column
df_q1=df_q1.drop(columns=['Query'])
df_q1
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Method', y='Execution Time [ms]', data=df_q1, palette=colors, hue='Method')

plt.title('Query 1')
plt.xlabel('Method')
plt.ylabel('Execution Time [ms]')
plt.show()
```
